Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing the dataset

```python
df = pd.read_csv("epl_20_21.csv")
df.head()
```

Checking the description of the numerical columns

```python
df.describe()
```

Checking for missing data

```python
df.isna().sum()
```

```python
#Adding two new columns
df["MinsPerMatch"] = df["Mins"]/df["Matches"].astype(int)
df["GoalsPerMatch"] = df["Goals"]/df["Matches"].astype(int)

df.head(3)
```

```python
#Total Goals 
Total_goals= df["Goals"].sum()
```

```python
#Total goals scored by penalties
Total_pk_goals = df["Penalty_Goals"].sum()
print(Total_pk_goals)
```

```python
#Total numbers of penalties attempted
Total_attempted_pk = df["Penalty_Attempted"].sum()
print(Total_attempted_pk)
```

```python
#Distribution between scored and missed penalties
plt.figure(figsize=(14,7))
pk_not_cvtd = Total_attempted_pk - Total_pk_goals
plt.title("Distribution of Penalties scored vs Missed", fontsize=20)
data = [pk_not_cvtd, Total_pk_goals]
labels = ["Missed Penalties", "Scored Penalties"]
color = sns.color_palette('Set2')
plt.pie(data, labels=labels, colors=color, autopct="%.0f%%")
plt.show()
```

### Checking Players with Multiple Positions

After checking for the unique player positions in the dataset, it is evident that there are some players that have multiple positions attached to their names. some of these multiple positions are Midfielder/Forward, Forward/Midfielder, Denfender/Midfielder, Midfielder/Defender, Defender/Midfielder. 

```python
#Players with multiple positions
df[df["Position"]== "MF,FW"]
```

After taking a closer at players with double positions, I came to the conclusion that some of these players have an second positional attachment due to where they were deployed by their managers during a particular game of the season. Using Kai Harvetz for example, he is originally a midfielder but due to the Chelsea's manager style of play, he was often deployed as a false striker. Other players could be deployed to different positions due to tactics or injuries. 

```python
df[df["Position"]== "FW,DF"]
```

This is a particurly interesting section because when I initially saw it I was shocked as to the abnormality of a Defender/Forward player. Then I investigated and found the players that fit this category. Using Bukayo Saka and Callum Hudson-Odoi as a case study, there are games where Bukayo Saka was deployed as a left back and Hudson-Odoi as a Right Winged Back and this could also be due to tactical reasons.

```python
df[df["Position"]== "DF,MF"]
```

There were 59 nationalities represented in the premier league for the 2020-21 season.

```python
nationality = df.groupby("Nationality").size().sort_values(ascending=False)
nationality.head(10).plot(kind="bar", figsize=(12,6), color=sns.color_palette("viridis"));
```

This chart shows the top 10 countries where majority of the players in the premier league come from with England being the highest nationality and Belgium being the lowest.

```python
df["Club"].value_counts().nlargest(5).plot(kind="bar", color=sns.color_palette("viridis"))
```

This barchart show the top five clubs with the highest squads for the 2020-21 season, West Bromwich Albion had the highest number with 30 players whilst Manchester United, Arsenal, Southampton and Everton had 29 players each

```python
df["Club"].value_counts().nsmallest(5).plot(kind="bar", color=sns.color_palette("viridis"))
```

This barchart is the opposite of the clubs with the highest squads, Leeds had the smallest squad in the 2020-21 season with about 23 players while the remaning clubs represented in the bar chart had about 24 players for the season

```python
#Players based on age bracket
Under20 = df[df["Age"] <=20]
btwn20_25 = df[(df["Age"] >20) & (df["Age"] <=25)]
btwn25_30 = df[(df["Age"] >25) & (df["Age"] <=30)]
Over30 = df[df["Age"] > 30]
```

```python
i = np.array([Under20["Name"].count(),btwn20_25["Name"].count(),btwn25_30["Name"].count(),Over30["Name"].count()])
labels = ["Under 20", "20-25","26-30", "Over 30"]
plt.figure(figsize=(12,6))
plt.title("Player distribution according to age", fontsize=20)
plt.pie(i, labels=labels, autopct="%.1f%%")
plt.show()
```

The chart above shows the percentage of players according their age in for the season.  The league was dominated by players in the 26-30 year old category with 37%  which was closely followed by players in the 20-25 age range with 35%. However, there was a low percentage of players above 30 years of age with 13.3%

```python
under_20 = df[df["Age"] < 20]
under_20["Club"].value_counts().plot(kind="bar", color=sns.color_palette("magma"))
```

The bar chart above shows Manchester United had the highest number of players below the age of 20 for the 2020-21 season and this was closely followed by Wolverhampton Wanders, then around the midpoint was Leicester City while Newcastle, Leeds, Tottenham, Burnley and Fulham had one player below the age of 20 each

```python
#Average age of players per club
plt.figure(figsize=(13,6))
sns.boxplot(x = "Club", y="Age", data = df)
plt.xticks(rotation=90);
```

From the boxplot, it is evident that Crystal Palace had the highest average age for the season and Manchest United had the lowest average age. There are also outliers at Cheslea and Sheffield United and these happens to be players above the age of 35. 

```python
avg_player_age = df.groupby("Club").size()
data = (df.groupby("Club")["Age"].sum())/ avg_player_age
data.sort_values(ascending=False)
```

```python
#Total assists from each club

club_assists = pd.DataFrame(df.groupby("Club", as_index=False)["Assists"].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x="Club", y="Assists", data=club_assists.sort_values(by="Assists"), palette="Set2")
ax.set_xlabel("Club")
ax.set_ylabel("Assists")
plt.xticks(rotation=90)
plt.rcParams["figure.figsize"]= (20,8)
plt.title("Club v Total Number of Assists", fontsize=20)
```

The barchart showes Manchester City has the highest number of goals coming from assists and this could be attributed to the style of play and other factors.

```python
#Top 10 assists

top_10_assists = df[["Name", "Club", "Assists", "Matches"]].nlargest(n=10, columns = "Assists")
top_10_assists
```

```python
#Total assists from each club

club_goals = pd.DataFrame(df.groupby("Club", as_index=False)["Goals"].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x="Club", y="Goals", data=club_goals.sort_values(by="Goals"), palette="Set1")
ax.set_xlabel("Club")
ax.set_ylabel("Goals")
plt.xticks(rotation=90)
plt.rcParams["figure.figsize"]= (20,8)
plt.title("Club v Total Number of Goals", fontsize=20);
```

The barchart shows the total number of goals scored by Manchester City which was just over 80 which is closely followed by Manchester United at 70 while Sheffield had the lowest at under 20

```python
#Top 10 Goals

top_10_goals = df[["Name", "Club", "Goals", "Matches"]].nlargest(n=10, columns = "Goals")
top_10_goals
```

```python
plt.figure(figsize=(15,8))
assists = df["Assists"].sum()
plt.title("Distribution of Goals Scored with or without assists", fontsize=18)
data = [Total_goals - assists, assists]
labels = ["Goals without assists", "Goals with Assist"]
color = sns.color_palette("Set1")
plt.pie(data, labels=labels, colors = color, autopct = "%.0f%%")
plt.show()
```

```python
yellow_cards = df.sort_values(by="Yellow_Cards", ascending=False)[:10]
plt.figure(figsize=(15,6))
plt.title("Player with most yellow cards", fontsize=20)
c = sns.barplot(x=yellow_cards["Name"], y=yellow_cards["Yellow_Cards"], label="Players", color="Yellow")
plt.ylabel("Number of Yellow Cards")
c.set_xticklabels(c.get_xticklabels(), rotation=90);
```

```python
red_cards = df.sort_values(by="Red_Cards", ascending=False)[:10]
plt.figure(figsize=(15,6))
plt.title("Player with most red cards", fontsize=20)
c = sns.barplot(x=red_cards["Name"], y=red_cards["Red_Cards"], label="Players", color="Red")
plt.ylabel("Number of Red Cards")
c.set_xticklabels(c.get_xticklabels(), rotation=90);
```
